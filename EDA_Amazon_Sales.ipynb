```python
# Part 2 - Data Cleaning
amazon_sales = data.dropna()
```

```python
amazon_sales.isnull().sum()
```

```python
amazon_sales.shape
```

```python
amazon_sales = amazon_sales.drop(['img_link','product_link'], axis=1)
amazon_sales.info()
```

```python
amazon_sales.head()
```

```python
amazon_sales = amazon_sales.drop(amazon_sales[amazon_sales['rating'] == '|'].index)
amazon_sales.loc[amazon_sales['rating'] == '|', ['rating', 'rating_count']]
amazon_sales.shape
```

```python
amazon_sales.discounted_price = amazon_sales.discounted_price.str.replace('₹','', regex=False).str.replace(',','', regex=False).astype('float64')
amazon_sales.actual_price = amazon_sales.actual_price.str.replace('₹','', regex=False).str.replace(',','', regex=False).astype('float64')
amazon_sales.discount_percentage = amazon_sales.discount_percentage.str.replace('%','', regex=False).astype('float64')
amazon_sales.rating = amazon_sales.rating.astype('float64')
amazon_sales.rating_count = amazon_sales.rating_count.str.replace(',','', regex=False).astype('int64')
amazon_sales.info()
```

```python
#Find the correlation between discount_percentage and rating
plt.figure(figsize=(10,6))
sns.scatterplot(data=amazon_sales, x='discount_percentage', y='rating', alpha=0.5)
sns.lmplot(data=amazon_sales, x='discount_percentage', y='rating', scatter=False)
plt.title('Correlation between Discount Percentage and Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.show()

#amazon_sales[['discount_percentage', 'rating']].corr()

correlation = amazon_sales['discount_percentage'].corr(amazon_sales['rating'])
print(f'Correlation between discount_percentage and rating: {correlation}')

#amazon_sales.plot.scatter(x='discount_percentage', y='rating', alpha=0.5)
```

```python
#Find the category with the most ratings count
amazon_sales.groupby('category').rating_count.sum().sort_values(ascending=False).head()
```

```python
#Find the most used word in product names inside the best category with the highest avg rating
from collections import Counter

#Step 1: Find the category with the highest average rating
top_categories = amazon_sales.groupby('category').rating.mean().idxmax()

#Step 2: Filter the products in that category and extract product names
top_products = amazon_sales[amazon_sales['category'] == top_categories]['product_name']

#Step 3: Split the product names into words and count the occurrences of each word
top_product_words = ' '.join(top_products).split()

#Step 4: Use Counter to find the most common words
Counter(top_product_words).most_common(10)
```

```python
#Find the correlation between len of words in about_product and rating_count
amazon_sales['about_product_len'] = amazon_sales['about_product'].apply(lambda x: len(x.split()))
plt.figure(figsize=(10,6))
sns.scatterplot(data=amazon_sales, x='about_product_len', y='rating_count', alpha=0.5)
sns.lmplot(data=amazon_sales, x='about_product_len', y='rating_count', scatter=False)
plt.title('Correlation between Length of About Product and Rating Count')
plt.xlabel('Length of About Product (in words)')
plt.ylabel('Rating Count')
plt.show()

correlation = amazon_sales['about_product_len'].corr(amazon_sales['rating_count'])
print(f'Correlation between about_product_len and rating_count: {correlation}')
```
